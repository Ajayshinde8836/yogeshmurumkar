```python
 from tkinter import *  # importing all libraries 

from timeit import default_timer as timer 

import random 


  
 # creating window using GUI 

window = Tk() 
window.title('Speed Typing test')

  
# the size of the window is defined 

window.geometry("600x300+400+100") 


x = 0


  
# defining the function for the test 

def game(): 

    global x 

  

    # loop for destroying the window 

    # after on test 

    if x == 0: 

        window.destroy() 

        x = x+1

  

    # defining function for results of test 

    def check_result(): 
        

        if entry.get() == words[word]: 
                
               

            # here start time is when the window 

            # is opened and end time is when 

            # window is destroyed 

            end = timer() 

            time= end-start
            
           

            # we deduct the start time from end 

            # time and calculate results using 

            # timeit function 

            print(time)
            
            print("Time taken for typing ‚è≤")
            print("greatüëç 100% Accuracy")
            print("wpm=7")
             
        else: 

            print("please write correct statment") 
            print("Accuracy=90%")
            print("wpm=15")
            

             
                  
    words = ['I love programming.',
             'coding is an art.',
             'I love India.', 
             'I learn python.',
             'we make a software.'] 

  

    # Give random words for testing the speed of user 

    word = random.randint(0, (len(words)-1)) 
    
    
 # start timer using timeit function 
     
    start = timer() 

    windows = Tk() 
    

    windows.geometry("600x300+400+100") 
    
    windows.title('Speed Typing Test')
    


  

    # use lable method of tkinter for labling in window 

    x2 = Label(windows, text=words[word], font="times 20") 

  

    # place of labling in window 

    x2.place(x=150, y=10) 

    x3 = Label(windows, text="Start Typing üíª", font="times 22") 

    x3.place(x=10, y=50)
    
    x4=Label(windows,text='Time:-',font=('airal',20,'italic bold'))
    x4.place(x=30,y=150)
    
    x5=Label(windows,text='Accuracy:-',font=('airal',20,'italic bold'))
    x5.place(x=30,y=200)
    
    x6=Label(windows,text='wpm:-',font=('airal',20,'italic bold'))
    x6.place(x=30,y=250)
    
    
  

    entry = Entry(windows,justify='center',font=('airal',15,'italic bold'))
    

    entry.place(x=200, y=60)
    
    
    entry.x4=Entry(windows,font=('airal',10,'italic bold'))
    entry.x4.place(x=174,y=160) 
   
    
    entry.x5=Entry(windows,font=('airal',10,'italic bold'))
    entry.x5.place(x=174,y=210) 
   
     
    entry.x6=Entry(windows,font=('airal',10,'italic bold'))
    entry.x6.place(x=174,y=260) 
   

    
    entry.focus_set()
   
  

    # buttons to submit output and check results 

    b2 = Button(windows, text="Done", 

                command=check_result, width=16, bg='grey') 

    b2.place(x=150, y=100) 

  

    b3 = Button(windows, text="Try Again",  

                command=game, width=16, bg='grey') 
    

    b3.place(x=250, y=100) 
    
     
    windows.mainloop() 

  

  

x1 = Label(window, text="Lets Check your Typing speed..üíªüòä", font="times 20") 

x1.place(x=10, y=50) 

  

b1 = Button(window, text="Go", command=game, width=16, bg='grey') 

b1.place(x=150, y=100) 

 
# calling window 
window.mainloop()

    

    
```
